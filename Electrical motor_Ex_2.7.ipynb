the motor, that is, the relation between speed and power. Show that maximum mechanical power is obtained at half the no-load speed and/or half the stall torque.
_Note: Power is the product of torque and angular velocity._

### Solution

Figure 2.25 shows the graph of a line, therefore, through the points indicated in it, we can find the equation of the form $y = mx + b$.

<p align="center">
<img src="DC_Motor.png", width=250, height=250>
</p>

Through this model of mechanical and electrical circuit, we will obtain two differential equations.

We apply the law of voltages of Kirchhoff to obtain the differential equation for the electric model.

$$L\frac{di(t)}{dt}+Ri(t) = V(t) - e     \hspace{1cm}(1)$$

Where $e$ represent the electromotive force:

$$e = K_{b}\dot{\theta}$$

To get the mechanical model, we have to consider Newton's second law for angular movement.

$$J\ddot{\theta} + b\dot{\theta} = K_{t}i(t) \hspace{1cm} (2)$$

Where $J$ and $b$ are respectively the moment of inertia of the rotor and the coefficient of damping by friction respectively.

Given that, the torque $T$ is:

$$T = K_{t}i(t)$$

With this, we have that the mechanical power in the rotor must equal the electric power, therefore:

$$T\dot{\theta} = ei(t)$$ 

$$K_{t}i(t)\dot{\theta} = i(t)K_{b}\dot{\theta}$$

$$K_{t} = K_{b}$$

Applying Laplace transform to equations (1) and (2):

$$(sL+R)I(s)=V(s)-K_{b}s\Theta(s) \hspace{1cm}(3)$$

$$s(sJ+b)\Theta(s)=K_{t}I(s) \hspace{1cm}(4)$$

Finally, by clearing $I(s)$ from equation (4) and replacing into equation (3), we can obtain an expression of the position-voltage transfer function:

$$\frac{\Theta(s)}{V(s)} = H(s) = \frac{K_{t}}{s((Js+b)(Ls+R)+K_{t}K_{b})} \hspace{1cm} \bigg[\frac{rad}{V}\bigg]$$

Now to demonstrate that the maximum mechanical power is obtained in the mean value of the 'Stall Torque' and/or in the mean value of the 'no-load speed', the equations are formulated to prove it mathematically.

Taking, $\tau_ {s}$, as the value for the 'stall torque' and, $\omega_ {n}$, as the value for the 'no-load speed'.

With these parameters defined, using the Torque-Velocity curve, we can obtain the equation of the form $ y = mx + b $; either torque based on angular velocity, or viceversa.

$$\tau_{motor} = -\Big(\frac{\tau_{s}}{\omega_{n}}\Big)\omega+\tau_{s} \hspace{1cm}(5)$$

$$\omega_{motor} = \frac{(\tau_{s}-\tau)\omega_{n}}{\tau_{s}} \hspace{1cm}(6)$$

Taking into account the note mentioned in the statement of the problem, the power is defined by:

$$P_{rot} = \tau\omega \hspace{1cm}(7)$$

You can replace the equation (5) or (6) into equation (7), it depends if you want to express the power as a function of the angular velocity or torque. Then, equation (5) is replaced in (7).

$$P_{motor}(\omega) = -\Big(\frac{\tau_{s}}{\omega_{n}}\Big)\omega^2+\tau_{s}\omega \hspace{1cm}(8)$$

Or expressing as a function of the torque:

$$P_{motor}(\tau) = -\Big(\frac{\omega_{n}}{\tau_{s}}\Big)\tau^2+\omega_{n}\tau \hspace{1cm}(9)$$

Equations (8) and (9), basically represent the same, the only difference is the variable which is depending the power. Therefore, when observing these equations with parabolic form, it can be derived and equal to zero to obtain the maximum power value.

As a function of angular velocity:

$$\frac{dP_{motor}(\omega)}{d\omega} = -2\Big(\frac{\tau_{s}}{\omega_{n}}\Big)\omega+\tau_{s} = 0$$

We obtain that:

$$P_{max} \longrightarrow \omega = \frac{\omega_{n}}{2}  $$

As a function of torque:

$$\frac{dP_{motor}(\tau)}{d\tau} = -2\Big(\frac{\omega_{n}}{\tau_{s}}\Big)\tau+\omega_{n} = 0$$

We obtain that:

$$P_{max} \longrightarrow \tau = \frac{\tau_{s}}{2}$$

In graphic form, we can use the algorithm below to visualize the shapes of the parabolas and their respective maximums.

```python
import matplotlib.pyplot as plt
from numpy import *
import numpy as np
from math import *
%matplotlib inline

ts = 0.05; wn = 6000;
w = np.arange(0,6000,1);
tl = np.arange(0,0.05,0.0001);
t = -(ts/wn)*w+ts;
plt.plot(w,t,'c',wn/2,ts/2,'ro');
plt.title('Graphic Torque vs. Speed');
plt.xlabel('Speed (rpm)');
plt.ylabel('Torque (N-m)');
plt.show();
```

Now we proceed to plot the parabolic functions.

**Power as a function of angular velocity:**

```python
P = -(ts/wn)*w**2+ts*w;
pn = -(ts/wn)*(wn/2)**2+ts*(wn/2);
plt.figure(1);
plt.subplot(211);
plt.plot(w,t,'c',wn/2,ts/2,'ro');
plt.title('Graphic Torque vs. Speed');
plt.xlabel('Speed (rpm)');
plt.ylabel('Torque (N-m)');
plt.subplot(212);
plt.plot(w,P,'g',wn/2,pn,'ro');
plt.title('Graphic Power vs. Speed');
plt.tight_layout();
plt.xlabel('Speed (rpm)');
plt.ylabel('Power');
plt.show();
```

**Power as a function of torque:**

```python
Pt = -(wn/ts)*tl**2+tl*wn;
ps = -(wn/ts)*(ts/2)**2+wn*(ts/2);
plt.figure(1);
plt.subplot(211);
plt.plot(w,t,wn/2,ts/2,'ro');
plt.title('Graphic Torque vs. Speed');
plt.xlabel('Speed (rpm)');
plt.ylabel('Torque (N-m)');
plt.subplot(212);
plt.plot(tl,Pt,'g',ts/2,ps,'ro');
plt.title('Graphic Potencia vs. Torque');
plt.tight_layout();
plt.xlabel('Torque (N-m)');
plt.ylabel('Power');
plt.show();
```

We can observe the points indicated in red, which in the lines correspond to the average values of both the 'stall torque' and the 'no-load speed', and these agree with the maximum in the parabola of the corresponding function.
